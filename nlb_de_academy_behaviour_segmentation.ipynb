**What is behavior segmentation?**

Behavior segmentation is a type of market segmentation based on differences in the behavior of different groups of costumers, taking into account their lifestyles, patterns of buying and using, patterns of spending money and time, demographic and  similar factors.

**Why do segmentation?**

Enables the bank to design a strategic marketing plan that more precisely meets the needs of the customer in a specific segment. Enhances bank’s focus, competitiveness, communication, retention and profitability.

*Clustering is a process of grouping a set of objects with similar characteristics into clusters. The main goal is to have objects in the groups as similar as possible, and as dissimilar as possible to the objects in the other group. Clustering is unsupervised learning technique, meaning that algorithm is looking for patterns within data, without a priori knowledge about the classification of data.*

# 3. Clustering (K-means clustering)

**K-means clustering** *(more information about method: https://scikit-learn.org/stable/modules/clustering.html#k-means)*

Required steps to perform K-means clustering:
1. Define initial number of clusters, k
2. Initially randomly select cluster centroids (centers)
3. Assign cases to closest centroid
4. Update cluster centroids by evaluating mean of the distances from each case to the centroid
5. Reassign cases to new centroids
6. Repeat steps 3 and 4 until convergence

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
# create dataset
X, y = make_blobs(
   n_samples=150, n_features=2,
   centers=6, cluster_std=0.5,
   shuffle=True, random_state=0
)

# plot
plt.scatter(
   X[:, 0], X[:, 1],
   c='white', marker='o',
   edgecolor='black', s=50
)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(
    n_clusters=6, init='random',
    n_init=10, max_iter=300, 
    tol=1e-04, random_state=0
)
y_km = km.fit_predict(X)
```

```python

# plot the 3 clusters
plt.scatter(
    X[y_km == 0, 0], X[y_km == 0, 1],
    s=50, c='lightgreen',
    marker='s', edgecolor='black',
    label='cluster 1'
)

plt.scatter(
    X[y_km == 1, 0], X[y_km == 1, 1],
    s=50, c='orange',
    marker='o', edgecolor='black',
    label='cluster 2'
)

plt.scatter(
    X[y_km == 2, 0], X[y_km == 2, 1],
    s=50, c='lightblue',
    marker='v', edgecolor='black',
    label='cluster 3'
)

# plot the centroids
plt.scatter(
    km.cluster_centers_[:, 0], km.cluster_centers_[:, 1],
    s=250, marker='*',
    c='red', edgecolor='black',
    label='centroids'
)
plt.legend(scatterpoints=1)
plt.grid()
plt.show()
```
